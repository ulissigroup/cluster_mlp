```python
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
```

```python
clus = ['Pd5', 'Cu6', 'Au8', 'Ni10', 'Cu13', 'Pd16', 'Cu4Au4', 'Ni6Pd4']
clus_subscript = ['Pd$_{5}$', 'Cu$_{6}$', 'Au$_{8}$', 'Ni$_{10}$', 'Cu$_{13}$', 'Pd$_{16}$', 'Cu$_{4}$Au$_{4}$', 'Ni$_{6}$Pd$_{4}$' ]
for item in clus:
    aa  = item+'_dist_pkl'
    aa = item + '/VASPCalls/' + item + 'similar_clus_diff_dist.pkl'
print(aa)
#dist_pkl = 'Pd5/VASPCalls/Pd5_similar_clus_diff_dist.pkl'
#ene_pkl =  'Pd5/VASPCalls/Pd5_similar_clus_ene_list.pkl'

#with open('Al7_similar_clus_ene_list.pkl', 'rb') as fh:
    #ene_list_pkl = pickle.load(fh)
```

```python
import numpy as np
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(36,15))

fig_order = [(0, 0), (0, 1), (0, 2), (0, 3), (1, 0), (1, 1), (1, 2), (1, 3)]

for i, item in enumerate(clus):
    print(item, fig_order[i])
    dist_pkl = item + '/VASPCalls/' + item + '_similar_clus_diff_dist.pkl'
    #ene_pkl =  item + '/VASPCalls/' + item + '_similar_clus_ene_list.pkl'
    print(dist_pkl)
    with open(dist_pkl, 'rb') as fh:
        clus_dist_diff = pickle.load(fh)
    print(len(clus_dist_diff))
    
    ax = axes[fig_order[i][0], fig_order[i][1]]
    #sns.set(font_scale=2)
    #sns.distplot( ax= ax, x=clus_dist_diff,  kde=True, color='g', kde_kws = {'color': 'r'})    
    sns.histplot( ax= ax, x=clus_dist_diff, binrange=(0,0.02),  kde=True, color='g') 
    ax.lines[0].set_color('crimson')
    ax.set_ylabel("Count", fontsize=16, weight='bold')
    #ax.set_xlabel("E(DFT) - E(Al)", fontsize=16, weight='bold')
    ax.set_title(clus_subscript[i], fontsize=24, weight='bold')
    ax.set_xticklabels(ax.get_xticks(), weight='bold', fontsize=16)
    ax.set_yticklabels(ax.get_yticks(), weight='bold', fontsize=16)
    #print("ax.get_xticks():", ax.get_xticks() )
    #tick_values = np.linspace(0.0, 0.02, num=5).tolist()
    #ax.set_xticklabels(tick_values,  fontsize=16)
    #ax.set_yticklabels(ax.get_yticks(),  fontsize=18)
    
plt.savefig("clus_hist.png")
```

```python
from matplotlib.ticker import FormatStrFormatter
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(30,12))

fig_order = [(0, 0), (0, 1), (0, 2), (0, 3), (1, 0), (1, 1), (1, 2), (1, 3)]

for i, item in enumerate(clus):
    print(item, fig_order[i])
    #dist_pkl = item + '/VASPCalls/' + item + '_similar_clus_diff_dist.pkl'
    ene_pkl =  item + '/VASPCalls/' + item + '_similar_clus_ene_list.pkl'
    print(ene_pkl)
    with open(ene_pkl, 'rb') as fh:
        clus_ene_diff = pickle.load(fh)
    print(len(clus_ene_diff))
    
    ax = axes[fig_order[i][0], fig_order[i][1]]
    #sns.distplot( ax= ax, x=clus_dist_diff,  kde=True, color='g', kde_kws = {'color': 'r'})    
    sns.histplot( ax= ax, x=clus_ene_diff,  binrange=(-0.02,0.02), kde=True, color='blue') 
    print("ax.get_xticks():", ax.get_xticks(),type(ax.get_xticks()) )
    #new_axis_lim = np.linspace(min(ax.get_xticks()), max(ax.get_xticks()), 5)
    new_axis_lim = np.linspace(-0.02, 0.02, 5)
    print("new_axis_lim", new_axis_lim)
    ax.lines[0].set_color('crimson')
    ax.set_title(clus_subscript[i], fontsize=16, weight='bold')
    ax.set_xticks(new_axis_lim)
    
    ax.set_xticklabels(ax.get_xticks(), weight='bold', fontsize=16)
    ax.set_yticklabels(ax.get_yticks(), weight='bold', fontsize=16)
    #ax.set_xticklabels(new_axis_lim, weight='bold', fontsize=16)
    ax.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))
    ax.set_ylabel("Count", fontsize=16, weight='bold')
    
    
plt.savefig("clus_ene.png")
```
